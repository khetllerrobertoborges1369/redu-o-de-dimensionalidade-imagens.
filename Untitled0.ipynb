```python
# ====================================
# Processamento de Imagens - Redução de Dimensionalidade
# Converter imagem Colorida -> Cinza -> Binária
# ====================================

# 1. Importar bibliotecas
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# 2. Upload da imagem
uploaded = files.upload()
nome_arquivo = list(uploaded.keys())[0]  # pega o nome da primeira imagem enviada

# 3. Ler a imagem colorida (cv2 lê em BGR)
imagem_colorida = cv2.imread(nome_arquivo)
imagem_colorida = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

# 4. Converter para escala de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_RGB2GRAY)

# 5. Binarizar (Threshold em 127)
_, imagem_binaria = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)

# 6. Mostrar resultados lado a lado
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.title("Colorida")
plt.imshow(imagem_colorida)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Cinza")
plt.imshow(imagem_cinza, cmap="gray")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Binária")
plt.imshow(imagem_binaria, cmap="gray")
plt.axis("off")

plt.show()
```
